```python
# bmi:weight(kg) / height(m)^2 程式設計

weight = float(input("請輸入體重: "))
height = float(input("請輸入身高: "))
bmi = weight / (height / 100) **2
print("你的體重是: " + str(weight))
print("你的身高是: " + str(height))
print("bmi是: " + str(bmi))

#>,>=,<,<=,==
if bmi > 25:    #這是一個篩選題, 最嚴格放上面, 只要找到第一個case就會結束所有判斷
    print ("過重")
elif bmi >18:
    print ("正常")
else:  #最後的分類, 後面沒有條件
    print ("過輕")
```

```python
# 終極密碼
import random

low, high = 1, 100
ans = random.randint(low+1, high-1)

while True:
    while True:
        print("請輸入", low, "-", high, ":")
        guess = int(input())
        if low < guess < high:
            break
        else:
            print("請重新輸入")
    if guess > low and guess < high:
        if guess > ans:
            print("猜大囉")
            high = guess
        elif guess < ans:
            print("猜小囉")
            low = guess
        else:
            print("恭喜猜對")
            break
```
