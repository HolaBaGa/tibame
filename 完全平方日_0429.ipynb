```python
# 定義函數何謂閏年
def is_leap_year(year):
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False
```

```python
# 定義函數判斷正確的年月日
# 20000831(O) 20001331(X) 20000229(O) 20010229(X)
# is_date_valid()
def is_date_valid(d):
    day = d % 100
    remain = d // 100
    month = remain % 100
    year = remain // 100
    if year >= 1:
        if 12 >= month >= 1:
            month_days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            max_days = month_days[month-1]
            # if is_leap_year(year) == True
            if is_leap_year(year) and month == 2:
                max_days = max_days + 1
            if max_days >= day >= 1:
                return True
            else:
                return False
        else:
            return False
    else:
        return False
print(is_date_valid(20000229))
print(is_date_valid(20010229))
```

```python
# 想法2 最快速能找到答案，運用+1平方的方式
# 20241001 2072...
start = 4499
result = []
while True:
    d = start * start
    if is_date_valid(d):
        # list專屬功能: append
        # l = l.append(X) l.append(O)
        result.append(d)
    if len(result) == 10:
        break
print(result)
start = start + 1
```

```python
# 想法2 最快速能找到答案
# 20241001 2072...
def fully_squared_2(start=4499, max_count=10):
    result = []
    while True:
        d = start * start
        if is_date_valid(d):
            # list專屬功能: append
            # l = l.append(X) l.append(O)
            result.append(d)
        if len(result) == max_count:
            break
        start = start + 1
    return result
print(fully_squared_2(max_count=5))
```

```python
# 想法1.從日期慢慢+1
import datetime as dt
d = dt.datetime(2024, 10, 1)
delta = dt.timedelta(days=1)
result = []
# 迴圈
while True:
    # 把日期往後一個一個加
    # 日期->字串->整數
    d_int = int(d.strftime("%Y%m%d"))
    # 開根號看是否為整數
    if (d_int ** 0.5).is_integer():
        # 把完全平方日日期轉成的整數放到result
        result.append(d_int)
    if len(result) == 10:
        break
    d = d + delta
print(result)
```

```python
def fully_squared_1(year=2024, month=10, day=1, max_count=10):
    d = dt.datetime(year, month, day)
    delta = dt.timedelta(days=1)
    result = []
    # 迴圈
    while True:
        # 把日期往後一個一個加
        # 日期->字串->整數
        d_int = int(d.strftime("%Y%m%d"))
        # 開根號看是否為整數
        if (d_int ** 0.5).is_integer():
            # 把完全平方日日期轉成的整數放到result
            result.append(d_int)
        if len(result) == max_count:
            break
        d = d + delta
    return result

fully_squared_1(max_count=20)
```

```python
# 在電腦裡算時間是使用秒數 Unix timestamp  使用格林威治時間
import time

total = 300

t1 = time.time()
fully_squared_1(max_count=total)
t2 = time.time()
dif1 = t2 - t1
print("想法1:", dif1, "秒")

t1 = time.time()
fully_squared_2(max_count=total)
t2 = time.time()
dif2 = t2 - t1
print("想法2:", dif2, "秒")

ratio = dif1 / dif2
print("相差", ratio, "倍")
```
