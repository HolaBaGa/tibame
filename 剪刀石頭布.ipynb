```python
# 按已知寫法製作, 若輸入超過012, 遊戲會錯誤

import random

win = 0
lose = 0
tw = "目前你獲勝的次數:"
tl = "目前你輸的次數:"

while True:
    my = int(input("請輸入:, 0剪刀 1石頭 2布"))
    com = random.randint(0, 2)
    trans = ["剪刀", "石頭", "布"]

    if my < 0 or my > 2:
        print("請重新輸入")

    if 0 <= my <= 2:
        print("我出", trans[my])
        print("電腦出", trans[com])

        if my == (com + 1) % 3:
            win = win + 1
            print("你贏了", tw, win, tl, lose)
            if win == 3:
                print("game over 你贏了")
                break

        elif com == (my + 1) % 3:
            lose = lose + 1
            print("你輸了", tw, win, tl, lose)
            if lose == 3:
                print("game over 你輸了")
                break

        else:
            print("平手")

```

```python
# 經過gemini 改良先以字串接收輸入並使用 .strip() 去除空格，接著用 .isdigit() 檢查是否為純數字。若輸入文字或空格，會印出錯誤提示並安全地重新詢問，程式不會崩潰。
# 在防呆檢查與範圍檢查中，都加入了 continue 關鍵字。一旦偵測到錯誤輸入，程式會立刻強制跳回迴圈開頭，不再往下讀取或執行無效的程式碼。
# 因為在範圍錯誤時（if my < 0 or my > 2:）就已經使用 continue 終止該輪迴圈，所以下方的遊戲邏輯不需要再被包在 if 0 <= my <= 2: 判斷式裡面，程式結構更扁平、易讀。
# 提示字串作了細微優化，在 input() 的最前面加上了換行符號 \n，且錯誤提示加上了 【輸入無效】 與 【範圍錯誤】 的視覺標籤，讓終端機的遊玩畫面更易於閱讀。

import random

win = 0
lose = 0
tw = "目前你獲勝的次數:"
tl = "目前你輸的次數:"

while True:
    # 1. 取得輸入並去除前後空格
    user_input = input("\n請輸入 0剪刀 1石頭 2布: ").strip()

    # 2. 防呆檢查：如果輸入為空、空格或非數字字元
    if user_input == "" or not user_input.isdigit():
        print("【輸入無效】請不要輸入空格或文字，請輸入數字 0, 1 或 2")
        continue # 跳回迴圈開頭重新輸入

    # 3. 確認是數字後才轉為 int
    my = int(user_input)
    com = random.randint(0, 2)
    trans = ["剪刀", "石頭", "布"]

    if my < 0 or my > 2:
        print("【範圍錯誤】請重新輸入 0 到 2 之間的數字")
        continue

    # 4. 遊戲邏輯
    print("我出", trans[my])
    print("電腦出", trans[com])

    if my == (com + 1) % 3:
        win = win + 1
        print("你贏了", tw, win, tl, lose)
        if win == 3:
            print("game over 你贏了")
            break

    elif com == (my + 1) % 3:
        lose = lose + 1
        print("你輸了", tw, win, tl, lose)
        if lose == 3:
            print("game over 你輸了")
            break

    else:
        print("平手")
```

```python
# 課程模擬操作

import random # 引用偽隨機數，統一放在最前面，只要是電腦做出的都是偽隨機數

my = int(input("請輸入 0.剪刀 1.石頭 2.布")) # 定義我出的, 清單一定從0開始, 特別是運用取餘數的公式, 用int 轉換成整數字, 限制輸入的型態
com = random.randint(0, 2) # 定義電腦出的, 隨機整數012
trans = ["剪刀", "石頭", "布"] # 定義字串, 讓數字能轉換成出中文
print("我出的: ", trans[my])
print("電腦出的: ", trans[com]) #語法會多一個空格 ","是分隔 "."是"的"的意思

if my == (com +1) %3:
    print("我贏了")
elif com == (my +1) %3:
    print("我輸了")
else:
    print("平手")
```

```python
#另一種剪刀石頭布
import random

win, lose = 0, 0
while win < 3 and lose < 3:
    #-1: lose, 0: evern, 1: win
    result = random.randint(-1, 1)
    print(result)
    if result == -1:
        lose = lose + 1
    elif result == 1:
        win = win + 1
if win > lose:
    print("WIN")
else:
    print("LOSE")
```

```python
#另一種剪刀石頭布 while True: break 語法 當判斷較多, 判斷位置較自由

win, lose = 0, 0
while True:
    #-1: lose, 0: evern, 1: win
    result = random.randint(-1, 1)
    print(result)
    if result == -1:
        lose = lose + 1
    elif result == 1:
        win = win + 1
    # 判斷
    if win == 3:
        print("WIN")
        break
    elif lose == 3:
        print("LOSE")
        break
```
