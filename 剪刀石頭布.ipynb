```python
import random

win = 0
lose = 0
tw = "目前你獲勝的次數:"
tl = "目前你輸的次數:"

while True:
    # 1. 取得輸入並去除前後空格
    user_input = input("\n請輸入 0剪刀 1石頭 2布: ").strip()

    # 2. 防呆檢查：如果輸入為空、空格或非數字字元
    if user_input == "" or not user_input.isdigit():
        print("【輸入無效】請不要輸入空格或文字，請輸入數字 0, 1 或 2")
        continue # 跳回迴圈開頭重新輸入

    # 3. 確認是數字後才轉為 int
    my = int(user_input)
    com = random.randint(0, 2)
    trans = ["剪刀", "石頭", "布"]

    if my < 0 or my > 2:
        print("【範圍錯誤】請重新輸入 0 到 2 之間的數字")
        continue

    # 4. 遊戲邏輯
    print("我出", trans[my])
    print("電腦出", trans[com])

    if my == (com + 1) % 3:
        win = win + 1
        print("你贏了", tw, win, tl, lose)
        if win == 3:
            print("game over 你贏了")
            break

    elif com == (my + 1) % 3:
        lose = lose + 1
        print("你輸了", tw, win, tl, lose)
        if lose == 3:
            print("game over 你輸了")
            break

    else:
        print("平手")
```
