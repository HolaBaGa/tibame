```python
# 多層迴圈(先寫裡面)
# 第一次
# hello
# hello
# 第二次
# hello
# hello
for i in range(3):
    print("第", i, "次")
    for j in range(2):
        print("hello")
```

```python
# 定義函式, 當作一層迴圈

def print2hello():
    for j in range(2):
        print("hello")

for i in range(3):
    print("第", i, "次")
    print2hello()
```

```python
#方法1

# oxoxo
# xoxox
# oxoxo
# xoxox

for i in range(4):
    line = ""
    for j in range(5):
        if i % 2 == 0:
            if j % 2 == 0:
                line = line + "o"
            else:
                line = line + "x"
        else:
            if j % 2 == 0:
                line = line + "x"
            else:
                line = line + "o"
    print(line)

```

```python
#方法2

# oxoxo
# xoxox
# oxoxo
# xoxox

for i in range(4):
    line = ""
    for j in range(5):
        # j = 0 2 4 6 8
        if (i + j) % 2 == 0:
            line = line + "o"
        # j = 1 3 5 7 9
        else:
            line = line + "x"
    print(line)
```

```python
# 進階寫法
symbols = ["o", "x", "@"]

#當i 循環5次
for i in range(5):
    #開始的位置是 i 除以 symbols的字串長度取餘數
    startpos = i % len(symbols)
    # line 是 一個字串
    line = ""
    # j 有5個迴圈
    for j in range(5):
    # 位置是 (起始位置 + j) 除以 字串長度取餘數
        pos = (startpos + j) % len(symbols)
    # line 的欄位是 line + 字串中的位置
        line = line + symbols[pos]
    #印出 line 清單
    print(line)
```
